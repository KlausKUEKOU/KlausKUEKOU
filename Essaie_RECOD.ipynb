```python
import nltk
import spacy
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
from wordcloud import WordCloud

# Télécharger les ressources nécessaires
nltk.download('stopwords')
nlp = spacy.load("fr_core_news_sm")  # Modèle français de spaCy

# Données : Réponses collectées
responses = [
    "Le gain de temps grâce à la suppression des trajets.",
    "Une meilleure concentration et productivité.",
    "Plus de flexibilité dans l'organisation du travail.",
    "Moins de stress lié aux transports.",
    "Équilibre entre vie professionnelle et vie personnelle amélioré.",
    "Flexibilité dans les horaires de travail."
]

# Nettoyage des textes
def preprocess_text(text):
    doc = nlp(text.lower())
    tokens = [token.lemma_ for token in doc if not token.is_stop and token.is_alpha]
    return " ".join(tokens)

cleaned_responses = [preprocess_text(resp) for resp in responses]

# Vectorisation TF-IDF
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(cleaned_responses)

# Clustering avec K-Means
num_clusters = 3  # Déterminer un nombre optimal selon le contexte
kmeans = KMeans(n_clusters=num_clusters, random_state=42, n_init=10)
kmeans.fit(X)
labels = kmeans.labels_

# Création d'un DataFrame pour l'affichage des résultats
df = pd.DataFrame({"Réponse": responses, "Cluster": labels})
print(df)

# Visualisation avec un nuage de mots
def generate_wordcloud(text_list):
    text = " ".join(text_list)
    wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)
    plt.figure(figsize=(10, 5))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis("off")
    plt.show()

generate_wordcloud(cleaned_responses)

# Affichage de la répartition des clusters
sns.countplot(x=df["Cluster"], palette="viridis")
plt.title("Répartition des thèmes identifiés")
plt.xlabel("Cluster")
plt.ylabel("Nombre de réponses")
plt.show()
```
